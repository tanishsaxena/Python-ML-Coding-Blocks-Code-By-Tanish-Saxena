```python
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator 
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("mnist_train_small.csv").values
```

```python
X=data[:,2:].reshape(-1,28,28,1) # One is single depth which conv takes
```

```python
y=data[:,1]
```

```python
datagen=ImageDataGenerator(horizontal_flip=True,
                           #vertical_flip=True,
                          zoom_range=2,
                          rotation_range=20,
                          shear_range=.3)
```

```python

gen=datagen.flow(X,y)
#l.append(gen)
```

```python
images,val=gen.next()
```

```python
images.shape
```

```python
val.shape
```

```python
for image,value in zip(images,val):
    plt.figure()
    plt.title(value)
    plt.style.use("seaborn")
    plt.imshow(image.reshape(28,28).astype(int),cmap="gray")
```

```python
gen_=datagen.flow_from_directory("datasets/")
```

```python
images_,values_=gen_.next()
```

```python
for image,values in zip(images_,values_):
    plt.figure()
    plt.title(np.argmax(values))
    plt.style.use("seaborn")
    plt.imshow(image.astype(int))
```
