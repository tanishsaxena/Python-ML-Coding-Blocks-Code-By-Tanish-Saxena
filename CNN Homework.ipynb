```python
filtr=np.array([[10,10,10],[0,0,0],[-10,-10,-10]])
```

```python
filtrrrr=filtr.T
```

```python
def filterr(img,filtr):
    res=np.zeros([img.shape[0]-2,img.shape[1]-2])
    for i in range(res.shape[0]):
        for j in range(res.shape[1]):
            cut=img[i:i+3,j:j+3]
            final=np.sum(cut*filtr)
            res[i][j]=final
    return res
```

```python
final_img_ver=filterr(gray,filtrrrr)
```

```python
final_img_hori=filterr(gray,filtr)
```

```python
plt.imshow(final_img_ver) ## Vertical edge detection
```

```python
plt.imshow(final_img_hori)  ## Horizontal edge detection
```
