```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
```

```python
data=pd.read_csv("mnist_train_small.csv").values
```

```python
X=data[:,2:]
y=data[:,1]
```

```python
pca=PCA()
```

```python
pca.fit(X)
```

```python
pca.components_.shape
```

```python
X_dash=np.dot(X,pca.components_.T)
```

```python
img=X_dash[0].reshape(28,28)
```

```python
plt.imshow(img)
```

```python
plt.plot(pca.explained_variance_)
```

```python
pca2=PCA(n_components=2)
```

```python
X_mod=pca2.fit_transform(X)
```

```python
X_select=X_mod[y<3]
y_select=y[y<3]
```

```python
plt.scatter(X_mod[:,0],X_mod[:,1],c=y)
```

```python
colors=["red","green","yellow","blue"]
```

```python
y_color=list(map(lambda x: colors[x],y_select))
```

```python
plt.scatter(X_select[:,0],X_select[:,1],c=y_color)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(5)
```

```python
model.fit(X_select,y_select)
```

```python
model.score(X_select,y_select)
```

```python
pca3=PCA(n_components=30)
```

```python
X_res=pca3.fit_transform(X)
```

```python
model2=KNeighborsClassifier(5)
```

```python
model2.fit(X_res,y)
```

```python
model2.score(X_res,y)
```
