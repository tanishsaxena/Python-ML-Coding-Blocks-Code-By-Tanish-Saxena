```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem import PorterStemmer,WordNetLemmatizer
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv("spam.csv",encoding="ISO-8859-1")
```

```python
df.head()
```

```python
X=df["v2"]
y=df["v1"]
```

```python
X.head()
```

```python
le=LabelEncoder()
```

```python
y_gen=le.fit_transform(y) # Encoding y
X_gen=list(X) # Converting X to a list
```

```python
y_gen
```

```python
X_gen
```

# Step 1 - Tokenizing all the sentences

```python
words=[]
for msg in X_gen:  
    words+=word_tokenize(msg.lower())
```

```python
contain=set(words)
```

```python
len(contain)
```

# Step 2 - Cleaning the words

```python
stop=stopwords.words("english")
punc=set(punctuation)
total=set(stop).union(set(punc)) 
pure=[w for w in contain if w not in total] #Pure words
```

```python
len(pure)
```

```python
port=PorterStemmer()
```

```python
much_pure=set(map(lambda w:port.stem(w),pure)) #Much Pure Words
```

```python
len(much_pure)
```

# Done above process with count vectorize

```python
cv=CountVectorizer(ngram_range=(1,2))
```

```python
vec=cv.fit_transform(X)
```

```python
vec=vec.todense()
```
