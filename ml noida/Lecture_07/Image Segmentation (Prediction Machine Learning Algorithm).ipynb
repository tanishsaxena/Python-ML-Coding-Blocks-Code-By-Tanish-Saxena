```python
from sklearn.datasets import make_blobs
import numpy as np
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
import random 
%matplotlib inline
import cv2
```

```python
img=cv2.imread("1.png")
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
X=img.reshape(-1,3)
```

```python
second=X.reshape(img.shape)
```

```python

X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(10)
```

```python
model.fit(X)
```

```python
color_pallets=model.cluster_centers_
```

```python
plt.figure()
plt.imshow(color_pallets.reshape(2,5,3)/255)
plt.show()
plt.figure()
plt.imshow(color_pallets.reshape(5,2,3)/255)
plt.show()
plt.figure()
#plt.imshow(color_pallets.reshape(3,3,3)/255)
#plt.show()
```
