```python
import cv2
import matplotlib.pyplot as plt
import pandas
import numpy as np
```

```python
img=cv2.imread("1.png")
```

```python
gray=(cv2.cvtColor(img,cv2.COLOR_RGB2GRAY))
```

```python
gray=cv2.resize(gray,(100,100))
```

```python
plt.imshow(gray)
```

```python
gray.shape
```

```python
filtr=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
filtr
```

```python
gray.shape
```

```python
def applyfilter(gray,filtr):
    res=np.zeros([gray.shape[0]-2,gray.shape[1]-2])
    for r in range(res.shape[0]):
        for c in range(res.shape[1]):
            cut=gray[r:r+3,c:c+3]
            res[r][c]=np.sum(cut*filtr)
    return res
```

```python
final_img=applyfilter(gray,filtr)
```

```python
final_img.shape
```

```python
plt.imshow(final_img,cmap="gray")
```

```python
filtr=np.array([[10,10,10],[10,10,10],[10,10,10]])
```

```python
imgg=applyfilter(gray,filtr)
```

```python
plt.imshow(imgg,cmap="gray")
```

```python
filtr_=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
def applyfilter_(gray,filtr):
    res=np.zeros([gray.shape[0]-(filtr.shape[0]-1),gray.shape[1]-(filtr.shape[1]-1)])
    for r in range(res.shape[0]):
        for c in range(res.shape[1]):
            cut=gray[r:r+(filtr.shape[0]),c:c+(filtr.shape[1])]
            res[r][c]=np.sum(cut*filtr)
    return res
```

```python
imggg=applyfilter_(gray,filtr_)
```

```python
plt.imshow(imggg,cmap="gray")
```

# Image Pooling

```python
def applypooling(gray,shape):
    res=np.zeros([gray.shape[0]//shape[0],gray.shape[1]//shape[1]])
    for r in range(0,res.shape[0]):
        for c in range(0,res.shape[1]):
            r_start=r*shape[0]
            c_start=c*shape[1]
            cut=gray[r_start:r_start+shape[0], c_start:c_start+shape[1]]
            res[r][c]=np.max(cut)

    return res
```

```python
pool=applypooling(gray,(1,1))
```

```python
plt.imshow(pool,cmap="gray")
```

```python
img=applyfilter_(pool,filtr_)
```

```python
plt.imshow(img,cmap="gray")
```
