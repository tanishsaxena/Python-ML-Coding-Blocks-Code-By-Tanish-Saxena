```python
x=np.linspace(0,2*np.pi,100)
si=np.sin(x)
co=np.cos(x)

plt.figure()
plt.plot(x,si,label="sin")
plt.plot(x,co,label="cos")
plt.xlabel("angles")
plt.ylabel("values")
plt.title("A Graph between sin and cos")
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
plt.legend()
plt.show


```
