```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression
from sklearn.datasets import make_blobs
```

This only works for two outputs more that two outputs it cannot divide

```python
X,y=make_blobs(1000,centers=2,random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()

model.fit(X,y)
```

```python
model.intercept_   # This is value of C in y=mx+c
```

```python
theta=model.coef_  # These are value of M in for 2 features of X i.e X1,X2 y=mx+c
```

```python
theta
```

```python
slope=-model.coef_[0][0]/model.coef_[0][1]
```

```python
intercept=-model.intercept_/model.coef_[0][1]
```

```python
x1_=X[:,0]
```

```python
x1=np.linspace(-8,8,5)
```

```python
x2=(slope *x1)+intercept 
```

```python
x2_=(slope *x1_)+intercept 
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(x1,x2)
plt.show()

plt.figure()
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(x1_,x2_)
plt.show()
```

```python
model.score(X,y)
```
