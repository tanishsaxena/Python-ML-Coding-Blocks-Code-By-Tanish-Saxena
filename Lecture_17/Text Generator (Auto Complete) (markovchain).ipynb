```python
text = open("speech.txt","r",encoding="utf8").read()
```

```python
import numpy as np
```

```python
model={}
```

```python
k=4
```

```python
for i in range(len(text)-k):
    cut=text[i:i+k]
    ch=text[i+k]
    if cut not in model:
        model[cut]={}
    if ch not in model[cut]:
        model[cut][ch]=1
    else:
        model[cut][ch]+=1
print(model)
```

```python
type(model[cut])
```

```python
generated="This is the purpose"
```

```python
for i in range(1000):
    cut=generated[-k:]
    freq=model[cut]
    freq_list=np.array(list(freq.items()))
    #Top three
    best_three=np.argsort(freq_list[:,1])[-3:]
    #generated random
    best=np.random.randint(0,len(best_three),1)[0]
    best_ch=freq_list[best,0]
    generated+=best_ch
```
