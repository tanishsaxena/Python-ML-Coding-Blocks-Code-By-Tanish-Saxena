# Self Made KMeans

```python
X,y=make_blobs(n_samples=1000,centers=2,random_state=0)
```

```python
class KMeans:
    
    def __init__(self,n_clusters,iters=10):
        self.n_clusters=n_clusters
        self.iters=iters
    
    def fit(self,X):
        self.X=X
        errors=[]
        
        points=random.choices(list(X),k=self.n_clusters) #Generating random n points as centers for n_clusters or n seed points
        
        # Create cluster corresponding to seed points
        self.clusters=[]
        for point in points:
            cluster=Cluster(point) #Created an object of Cluster class
            self.clusters.append(cluster)
            
        for _ in range(self.iters): # Running full loop again again for accuracy
            
            self.labels_=[]
            for point in X:
                distances=[]
                for cluster in self.clusters:
                    dis=cluster.distance(point)
                    distances.append((dis,cluster))
                distances_np=np.array(distances) # Converting list of all the distances of a point from all centers
                
                min_index=np.argmin(distances_np[:,0].astype(int))
                selected_cluster=self.clusters[min_index]
                selected_cluster.points.append(point)
                self.labels_.append(min_index)
                
            for cluster in self.clusters:
                cluster.update_center()
                
            errors.append(self.error)
        return errors
    
    def error(self):
        total=0
        for point, label in zip(self.X,self.labels_):
            dis=self.clusters[label].distance(point)
            total+=dis
        return total
        
            
            
class Cluster:
    
    def __init__(self,center):
        self.center=center
        self.points=[]
        
    def update_center(self):
        self.center = np.mean(self.points,axis=0)
        self.points=[]
        
    def distance(self,point):
        return np.sum((self.center-point)**2)
    
    def __repr__ (self):
        return self.cluster
    
```
